```python
from math import *
def haversine(koord1,koord2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [koord1[0],koord1[1],koord2[0],koord2[1]])
    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    # Radius of earth in kilometers is 6371
    dist = 6371000 * c
    return dist
```

```python
from dataclasses import dataclass
```

```python
from math import sqrt, sin, cos
from typing import Union


@dataclass
class Vector:
    x: float
    y: float
    
    def __add__(self, other):
        return Vector(self.x+other.x, self.y+other.y)
    
    def __neg__(self):
        return Vector(-self.x, -self.y)
    
    def __sub__(self, other):
        return self + (-other)
    
    def __abs__(self):
        return sqrt(self.x**2 + self.y**2)
    
    def __mul__(self, other):
        return self.x * other.x + self.y * other.y
    
    def rotate(self, fi: float):
        x, y = self.x, self.y
        self.x = x * cos(fi) - y * sin(fi)
        self.y = x * sin(fi) + y * cos(fi)
```

```python
from math import *
def haversine(koord1: Vector, koord2: Vector) -> float:
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [koord1.x, koord1.y, koord2.x, koord2.y])
    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    # Radius of earth in kilometers is 6371
    dist = 6371000 * c
    return dist

def get_angle(a: Vector, b: Vector)-> float:
    return acos(a * b / (abs(a) * abs(b)))

def sign(a)->int:
    return -1 if a < 0 else 1

def elipse_generator(a: Vector, b: Vector, r: float, points_num: int=10):
    l = r / 2
    s = sqrt(r**2 - abs(b - a)**2)/2
    angle = get_angle(b - a, Vector(1, 0))
    
    if (b - a).y < 0:
        angle = 2 * pi - angle
    
    for i in range(points_num):
        t = 2 * pi / points_num * i
        x = l * sin(t)
        y = s * cos(t)
        v = Vector(x, y)
        v.rotate(angle)
        v += Vector((a + b).x / 2, (a + b).y / 2)
        yield v
        
```

```python
a = Vector(1., 3.)
b = Vector(3., 1.)
points = list(elipse_generator(a, b, 10, 10))
```

```python
from functools import reduce

def get_distance(a_id: int, b_id: int)-> float:
    a_id = int(a_id)
    b_id = int(b_id)
    a = Vector(float(nodes[a_id]['lat']), float(nodes[a_id]['lon']))
    b = Vector(float(nodes[b_id]['lat']), float(nodes[b_id]['lon']))
    return abs((a - b))

def findWay(a: Vector, b: Vector, l: float):
    way = []
    m = 100000
    actual_l = l / 111_139
    for start_id in nbrs.kneighbors([(a.x, a.y)], return_distance=False)[0]:
        for end_id in nbrs.kneighbors([(b.x, b.y)], return_distance=False)[0]:
            for middle in elipse_generator(Vector(*node_vals[start_id]), Vector(*node_vals[end_id]), actual_l, 10):
                for middle_id in nbrs.kneighbors([(middle.x, middle.y)], return_distance=False)[0]:
                    
                    start = str(node_ids[start_id])
                    end = str(node_ids[end_id])
                    middle = str(node_ids[middle_id])
                    
                    current_way = nx.algorithms.shortest_paths.shortest_path(g, start, middle) + \
                                  nx.algorithms.shortest_paths.shortest_path(g, middle, end)
                    current_len = sum(map(lambda x: get_distance(*x), zip(current_way[:-1], current_way[1:])))

                    if abs(current_len  - actual_l) < m:
                        m = abs(current_len - actual_l)
                        way = current_way
    return None if not way else way
```
